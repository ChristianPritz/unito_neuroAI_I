## Loading the dataset and partitioning into training, validation and test datasets 


```python
# loading 

DATA_DIR = "unito_neuroAI_I/datasets"
BATCH_SIZE = 32
IMG_SIZE = (128, 128)
SEED = 42

full_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    image_size=IMG_SIZE,
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

dataset_size = tf.data.experimental.cardinality(full_ds).numpy()
print("Total batches:", dataset_size)

# Partitioning 

train_frac = 0.7
val_frac   = 0.15
test_frac  = 0.15

train_batches = int(train_frac * dataset_size)
val_batches   = int(val_frac * dataset_size)

train_ds = full_ds.take(train_batches)
remaining = full_ds.skip(train_batches)

val_ds = remaining.take(val_batches)
test_ds = remaining.skip(val_batches)

# Autotune 

AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)
test_ds  = test_ds.cache().prefetch(AUTOTUNE)
```

### Getting to know the data

```python
class_names = full_ds.class_names
num_classes = len(class_names)

print("Classes:", num_classes)
print(class_names[:10])

images, labels = next(iter(train_ds))
print("Shape of the image tensor:", images.shape)
print("Shape of the label vector:", labels.shape)

images, labels = next(iter(train_ds))
input_shape = images.shape[1:]




# Get one batch
images, labels = next(iter(train_ds))

plt.figure(figsize=(8, 8))
for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    img = images[i].numpy().squeeze()  # remove channel dim (H,W,1)  (H,W)
    plt.imshow(img, cmap="gray")
    plt.title(class_names[labels[i]])
    plt.axis("off")

plt.tight_layout()
plt.show()

```

## Defining the model architecture

*CNN* with 3 convolutional and 2 dense layers

![CNN architecture](architecture.png)

```python
model = models.Sequential([
    layers.Input(shape=input_shape),
    
    layers.Conv2D(16, 3, activation='relu', padding='same'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(32, 3, activation='relu', padding='same'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(64, 3, activation='relu', padding='same'),
    layers.MaxPooling2D(),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

## Training the model over 15 Epochs


```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15
)
```

## Testing the trained network on a few instances ...

```python

# Take one batch from test dataset
images, labels = next(iter(test_ds))

# Run inference
preds = model.predict(images)
pred_classes = tf.argmax(preds, axis=1)

# Plot a few examples
n_show = 8
plt.figure(figsize=(12, 6))

for i in range(n_show):
    ax = plt.subplot(2, 4, i + 1)

    # Remove channel dimension for grayscale
    img = images[i].numpy().squeeze()

    plt.imshow(img, cmap="gray")
    plt.axis("off")

    true_label = class_names[labels[i]]
    pred_label = class_names[pred_classes[i]]

    color = "green" if true_label == pred_label else "red"
    plt.title(f"P: {pred_label}\nT: {true_label}", color=color)

plt.tight_layout()
plt.show()
```

## Doing inference on the entire test set. 

```python
test_loss, test_acc = model.evaluate(test_ds, verbose=1)

print(f"Test accuracy: {test_acc:.4f}")
```

## Visualizing the convolutional filters

```python

filters, biases = model.layers[0].get_weights()

plt.figure(figsize=(6,6))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(filters[:,:,0,i], cmap='gray')
    plt.axis('off')
plt.suptitle("First-layer filters (edges)")
plt.show()
```

## Outputs from the convolutional layers



```python
# Build activation model
images, labels = next(iter(train_ds))

# Force build
_ = model(images)

activation_model = tf.keras.Model(
    inputs=model.inputs,
    outputs=[layer.output for layer in model.layers
             if isinstance(layer, tf.keras.layers.Conv2D)]
)

activations = activation_model(images[:1])

for layer_idx, act in enumerate(activations):
    plt.figure(figsize=(6,6))
    for i in range(min(16, act.shape[-1])):
        plt.subplot(4,4,i+1)
        plt.imshow(act[0,:,:,i], cmap='gray')
        plt.axis('off')
    plt.suptitle(f"Conv layer {layer_idx+1}")
    plt.show()
```

## Display the embeddings in the dense layers

```python
images, labels = next(iter(train_ds))

# Force build
_ = model(images)
embedding_model = tf.keras.Model(
    inputs=model.inputs,
    outputs=model.layers[-2].output   # dense_2
)

embeddings = embedding_model.predict(images)
print(embeddings.shape)
    


pca = PCA(n_components=2)
Z = pca.fit_transform(embeddings)

plt.figure(figsize=(6,6))
for i, name in enumerate(class_names):
    idx = labels == i
    plt.scatter(Z[idx,0], Z[idx,1], label=name)
plt.legend()
plt.title("Face embeddings (CNN penultimate layer)")
plt.show()
```
